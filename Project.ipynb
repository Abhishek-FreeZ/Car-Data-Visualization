```python
import pandas as pd
import plotly.express as px
import plotly.offline as pyo
import plotly.graph_objs as go
pyo.init_notebook_mode(connected=True)
```

```python
brandModelPrice = pd.read_csv("Outputs/brandModelPrice.csv")
brandPrice= pd.read_csv("Outputs/brandPrice.csv")
fuelTypePrice= pd.read_csv("Outputs/fuelTypePrice.csv")
fuelTypeBrandPrice= pd.read_csv("Outputs/fuelTypeBrandPrice.csv")
yearWiseBrandPrice= pd.read_csv("Outputs/yearWiseBrandPrice.csv")
transmissionType = pd.read_csv("Outputs/transmissionType.csv")
mileageData=pd.read_csv("Outputs/mileageData.csv")
yearlyTransmission=pd.read_csv("Outputs/yearlyTransmission.csv")
```

```python
fig1 = px.bar(
    data_frame = brandPrice,
    x='brand',
    y='Average Price',
    color='brand',
    text='Average Price',
    hover_name='brand',
#     hover_data=['Min Price','Max Price'],
    hover_data={
        'brand':False,
        'Min Price':True,
        'Max Price':True
    },
#     opacity=0.9,
#     orientation='v',
    title='Brand Wide Average Price',
    labels={
        'brand':'Brand'      
    },
    template='plotly_dark',
#     width=1280,
#     height=720,
#     barmode='group',
#     facet_col='brand',
#     facet_row='brand'
#     facet_col_wrap=2
)
fig1.update_layout(
    hoverlabel=dict(
#         bgcolor='white',
        font_size=16,
        font_family="Rockwell"
    )
)
```
